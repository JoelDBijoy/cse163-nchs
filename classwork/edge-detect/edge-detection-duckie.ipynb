````python
import imageio as iio
import matplotlib.pyplot as plt
import numpy as np

# Write your function here!
def edge_detect(image_file: str) -> None:
    # Define the kernel
    
    # Apply the kernel
    
    # Normalize the image
    
    img = iio.imread(image_file, mode = "L")
    



  
    kernel_x = np.array([[-1, 0, 1],
                         [-2, 0, 2],
                         [-1, 0, 1]])
    
   
    kernel_y = np.array([[-1, -2, -1],
                         [ 0,  0,  0],
                         [ 1,  2,  1]])


    rows, cols = img.shape
    

    edges_x = np.zeros((rows - 2, cols - 2))
    edges_y = np.zeros((rows - 2, cols - 2))

    for i in range(1, rows - 1):
        for j in range(1, cols - 1):
            neighborhood = img[i-1:i+2, j-1:j+2]
            edges_x[i-1, j-1] = np.sum(neighborhood * kernel_x)
            edges_y[i-1, j-1] = np.sum(neighborhood * kernel_y)

    edge_magnitude = np.sqrt(edges_x**2 + edges_y**2)


    normalized_edges = (edge_magnitude * 255.0 / edge_magnitude.max())
 
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray' if img.ndim == 2 else None)

    plt.subplot(1, 2, 2)
   
    plt.imshow(normalized_edges, cmap='gray')
    

    plt.show()

    
edge_detect("duck.jpg")
````
